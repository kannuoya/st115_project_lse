```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
# list of names of department of economics faulty staff from London School of Economics and Political Science

# Send a request to the URL of the web page
page = requests.get("https://www.lse.ac.uk/economics/people/faculty")

# Create a BeautifulSoup object from the page content
lse_soup = BeautifulSoup(page.content, 'html.parser')
lse_divs = lse_soup.select('html > body > form > div:nth-of-type(3) > div > div:nth-of-type(2) > div > div')
lse_num_of_staff = len(lse_divs) + 1
lse_names = []

for n in range(1,lse_num_of_staff):
    lse_staff = lse_soup.select_one(f'html > body > form > div:nth-of-type(3) > div > div:nth-of-type(2) > div > div:nth-of-type({n}) > div:nth-of-type(2) > p > a')
    lse_staff_detail = lse_staff.text
    # Split the string into two parts at the first occurrence of "/"
    parts = lse_staff_detail.split('/', 1)
    # Get the first part of the resulting list for the full name
    substring = parts[0]
    lse_names.append(substring)
```

```python
# list of names of department of economics faulty staff from University College of London

ucl_url = 'https://www.ucl.ac.uk/economics/people/academic-and-teaching-staff'
ucl_r = requests.get(ucl_url)
ucl_soup = BeautifulSoup(ucl_r.content,'lxml')
ucl_names = []
for tab in ucl_soup.find_all('article'):
    for section in tab.find_all('section', {'class': 'middle-split__column1'}):
        for h4 in section.find('h4'):
            # clean the name
            h4 = (h4.text).replace('\xa0','')
            if h4 == '':
                break
            else:
                ucl_names.append(h4)
```

## 3 Data preparation

### Data cleaning importing and exploration

```python
def extract_area(name):
    
    '''
    This function extracts the part of response including subjects of academic work by a particular faculty member
    '''
    
    surname = name.split(' ')[1:]
    first_name = name.split(' ')[0]
    url = f'https://api.springernature.com/metadata/json?api_key={api_key}&q=name%3A%22{surname}%2C%20{first_name}%22%2C%20subject%3AEconomics&s=1&p=10'
 
    return json.loads(requests.get(url).text)['facets'][0]['values']

def extract_type(name):
    
    '''
    This funciton extracts the part of the response including types of academic work by a particular faculty member
    '''
    
    surname = name.split(' ')[1:]
    first_name = name.split(' ')[0]
    url = f'https://api.springernature.com/metadata/json?api_key={api_key}&q=name%3A%22{surname}%2C%20{first_name}%22%2C%20subject%3AEconomics&s=1&p=10'
    return json.loads(requests.get(url).text)['facets'][-1]['values']
     
def get_research_area_count(names):
    
    '''
    This function convert a list of names to a dictionary a dictionary which keys are research areas by all the members and values are their counts
    names: a list of strings
    return: a dictionary which keys are research areas by all the members and values are their counts
    '''
    
    research_area = {}
    for name in names: 
        list_of_dicts = extract_area(name) 
        for d in list_of_dicts:
            area = d['value'] 
            count = int(d['count'])
            if area not in research_area.keys():
                research_area[area] = count
            else:
                research_area[area] += count
    return research_area

def get_professor_paper_count(names):
    
    '''
    This function convert a list of names to a dictionary which keys are names of all the members and values are number of academic papers/books/journals they have done
    names: a list of strings
    return: a dictionary which keys are names of all the members and values are number of academic papers/books/journals they have done
    '''
    
    paper_count = {}
    for name in names:
        list_of_dicts = extract_type(name) 
        count = 0
        for d in list_of_dicts:
            count += int(d['count']) 
        paper_count[name] = count 
        if paper_count[name] == 0:
            paper_count.pop(name)
    return paper_count
             
```

```python
def get_research_area_dataframe(names):
    
    '''
    This function provides information about research areas in the form of pandas dataframe
    '''
    
    df = pd.DataFrame()
    df['research area'] = list(get_research_area_count(names).keys())
    df['total number'] = list(get_research_area_count(names).values())
    sum_of_papers = sum(list(get_professor_paper_count(names).values()))
    df['proportion of the area'] = [x/sum_of_papers for x in df['total number']]
    df = df.sort_values('total number', ascending = False).reset_index(drop = True)
    return df
```

All the above DataFrames are storing each university's research area, and its respective times of each area is researched and proportion to the total number of paper. Then based on all those we can get the average number of research paper that is published by each university.

```python
def get_average(names):
    
    '''
    #This function calculates average amount of academic work per teaching staff
    '''
    
    paper_count = get_professor_paper_count(names)
    average = sum(list(paper_count.values())) / len(paper_count)
    return average
```

```python
universities = ['Oxford', 'LSE', 'UCL', 'Cambridge']
all_names = [ox_names, lse_names, ucl_names, cam_names]
uni_avr = []

print("##Average amount of academic work per teaching staff in Department of Economics")
print("---------------------------------------------------------------------------------")

for i in range(len(universities)):
    one_average = get_average(all_names[i])
    uni_avr.append(one_average)
    print(f"{universities[i]}: {one_average}")
```

```python
import numpy as np


def plot_research_areas(uni_list, str_names):
    """
    Plot pie charts for the top 5 research areas of each department in the given list of department names.

    Parameters:
    -----------
    uni_list : list of DataFrame
        A list of universities' names for which to plot the research areas.
    str_names: list of str
    Returns:
    --------
    None
    """
    # Create a 2x2 grid of subplots for each department's pie chart
    fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(20, 20))
    axs = axs.flatten()

    # Loop over each department and plot a pie chart for the top 5 research areas
    for i, uni_df in enumerate(uni_list):
        top_five_research_area = uni_df.iloc[1:6]
        sizes = top_five_research_area['proportion of the area']
        labels = top_five_research_area['research area']
        colors = plt.cm.Blues_r(np.linspace(0.3, 1, len(sizes)))
        explode = [0.2] + [0] * (len(sizes) - 1) # explode the largest slice
        axs[i].pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, explode=explode)
        axs[i].set_title(f'Main research area of {str_names[i]} Economics Department', fontsize=20, fontweight='bold')

    # Add a common legend for all pie charts
    handles, labels = axs[0].get_legend_handles_labels()
    fig.legend(handles, labels, loc='upper left')

    fig.subplots_adjust(wspace=0.5)
    plt.show()

df_all_names = [df_ox, df_lse, df_ucl, df_cam]
plot_research_areas(df_all_names, universities)
```

This combined group of pie charts was utilized to ascertain the predominant areas of expertise within the economics departments of top United Kingdom universities. The findings revealed that Public Finance is the primary strength of Oxford, while Game Theory, Economics, and Social and Behavioral Sciences are the focal areas at University College London (UCL). Notably, London School of Economics (LSE) and Cambridge both specialize in general Economics.

**University of Cambridge**: Cambridge has a long and distinguished history of producing influential economists and nurturing ground-breaking economic theories, such as the development of Keynesian economics. **University of Oxford**: Oxford's expertise in Public Finance can be attributed to its long-standing reputation in economics, politics, and social sciences. Additionally, Oxford's emphasis on interdisciplinary research allows scholars to address complex issues in Public Finance from multiple perspectives.

**University College London (UCL)**: UCL's focus on Game Theory, Economics, and Social and Behavioral Sciences is influenced by its historical commitment to innovative research and cutting-edge methodologies. The university's interdisciplinary approach encourages scholars to apply Game Theory insights to various fields, fostering the growth and development of this area of focus. Moreover, UCL's location in London provides unique opportunities for collaboration with leading institutions and organizations, enriching its research in these disciplines.

In conclusion, while LSE, Cambridge, Oxford, and UCL share similarities in their focus on Economics, Public Finance, and Social and Behavioral Sciences, each institution possesses distinct factors that contribute to their unique specializations and strengths. These factors include historical context, intellectual legacy, location, and institutional priorities, which shape the research agendas and academic offerings at each university.
